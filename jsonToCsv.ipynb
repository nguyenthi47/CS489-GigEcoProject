```python
import json
import pandas as pd
```

```python
with open("survey_data/nov30.json") as fd:
    tmp = json.load(fd)
len(tmp), tmp
```

```python
cols = ["age", "consentEnd", "demographicsEnd", "education", "gender", "gigBenefit", "gigEver", "gigWorkers", 
       "income", "introEnd", "politics", "race", "studyEnd"] + list(map(lambda x: "scenario{}end".format(str(x)), range(0, 9)))
work_cols = ['Day off', "Direct Supervisor", "Evaluation", "Salary", "Working Options (companies in same sector)",
             "Working hours (length)", "Working hours (time period)"]
```

```python
long_responses = []
error_pid = []
for pid in tmp:
    for i in range(0, 10):
        for j in range(0, 2):
            try:
                print("pid: {}, comparison: {}, job: {}".format(pid, i, j + 1))
                response = {id_: tmp[pid][id_] for id_ in cols}
                response.update({id_: tmp[pid]["pairs"][i][j][id_] for id_ in work_cols})
                response["choice"] = tmp[pid]["choice" + str(i)] == j + 1
                response["eval"] = tmp[pid]["j" + str(i) + "-" + str(j + 1)]
                response["pid"] = pid
                response["rid"] = i
                long_responses.append(response)
            except:
                error_pid.append(pid)
                print("Error: {}".format(pid))
```

```python
long_responses = list(filter(lambda x: x["pid"] not in error_pid, long_responses))
```
