```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df.csv')
df.head()
```

```python
conjoint_data = pd.get_dummies(df,columns=['Day off', 'Direct Supervisor','Evaluation','Salary','Working Options (companies in same sector)','Working hours (length)','Working hours (time period)'])
conjoint_data.head()
```

```python
X = conjoint_data[['Day off_Paid, but following company policies',
                  'Day off_Unpaid, but whenever you want',
                  'Direct Supervisor_An algorithm',
                  'Direct Supervisor_No',
                  'Direct Supervisor_Yes',
                  'Evaluation_Daily',
                  'Evaluation_Monthly',
                  'Evaluation_Yearly',
                  'Salary_Fixed US$250.00 + US$6.25 per hour',
                  'Salary_Fixed US$500.00',
                  'Salary_US$12.50 per hour',
                  'Working Options (companies in same sector)_No',
                  'Working Options (companies in same sector)_Yes',
                  'Working hours (length)_Minimum 4h',
                  'Working hours (length)_Minimum 8h',
                  'Working hours (length)_Work as much as you want/need',
                  'Working hours (time period)_9am-5pm required',
                  'Working hours (time period)_Totally flexible']]
X = sm.add_constant(X)
df['choice'] = df['choice'].astype(int)
Y = df['choice']
```

```python
linearRegression = sm.OLS(Y, X).fit()
linearRegression.summary()
```

```python
attributes = ['Day off_Paid, but following company policies',
                  'Day off_Unpaid, but whenever you want',
                  'Direct Supervisor_An algorithm',
                  'Direct Supervisor_No',
                  'Direct Supervisor_Yes',
                  'Evaluation_Daily',
                  'Evaluation_Monthly',
                  'Evaluation_Yearly',
                  'Salary_Fixed US$250.00 + US$6.25 per hour',
                  'Salary_Fixed US$500.00',
                  'Salary_US$12.50 per hour',
                  'Working Options (companies in same sector)_No',
                  'Working Options (companies in same sector)_Yes',
                  'Working hours (length)_Minimum 4h',
                  'Working hours (length)_Minimum 8h',
                  'Working hours (length)_Work as much as you want/need',
                  'Working hours (time period)_9am-5pm required',
                  'Working hours (time period)_Totally flexible']
part_worth  = []
part_worth_range = []
level_name = []
end = 1
for item in attributes:
    nlevels = len(list(set(conjoint_data[item])))
    level_name.append(list(set(conjoint_data[item])))
    begin = end
    end = begin + nlevels - 1
    new_part_worth = list(linearRegression.params[begin:end])
    new_part_worth.append((-1) * sum(new_part_worth))
    part_worth_range.append(max(new_part_worth) - min(new_part_worth))
    part_worth.append(new_part_worth)

attribute_importance = []
for item in part_worth_range:
    attribute_importance.append(round(100 * (item / sum(part_worth_range)),2))
    
estimates_of_choice = []
index = 0 
for item in attributes : 
    print ("\nAttribute : " , item)
    print ("Importance : " , attribute_importance[index])
#     print('Level Part-Worths')
#     for level in range(len(level_name[index])):
#         print(level_name[index][level], part_worth[index][level])
    index = index + 1
```
